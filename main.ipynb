3. Calculate Angles

```python
def calculate_angle(a,b,c):
    a=np.array(a)   #first(shoulder)
    b=np.array(b)   #second(elbow)
    c=np.array(c)   #end(wrist)

    radians = np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])    #calculating angle between shoilder, elbow and wrist
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle    #Hand can't have angle more than 180
    return angle
```

4. Curl Counter

```python
'''
We are going test our feed from our web cam (Video Feed)
'''
cap=cv2.VideoCapture(0) # it is used to turn on the vidqeo capturing device and the number 0 is assigned to video capturing device

#Curl Counter Variables
counter=0
stage=None


with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose: #accuracy of our tracker is set (if we set more accuracy then detection power will decrease and vice versa)

    while cap.isOpened():
        ret, frame = cap.read() #ret is a variable which is used to return. Values of curls 
                            #frame will return image from our webcam 
        #Recolor image to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)   #coloring the frame to render in an array of colors
        image.flags.writeable= False    #setting the frame whether to take it or not

        #Make detection
        results= pose.process(image)  

        #Recolor back to BGR
        image.flags.writeable = True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)  #bring back the original image


        #Extract landmarks
        try:
            landmarks=results.pose_landmarks.landmark

            #Get Co-ordinates
            shoulder=[landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow=[landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist=[landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]  

            #calculate Angle  
            angle=calculate_angle(shoulder,elbow,wrist)
            
            #Visualize(in screen)
            cv2.putText(image,str(angle),
                            tuple(np.multiply(elbow, [640,480]).astype(int)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255),2,cv2.LINE_AA )
            

            #Curl Counter Logic
            if angle>160:
                stage="down"
            if angle<30 and stage == 'down':
                stage="up"
                counter+=1
                print(counter)


        except:
            pass 

        #Render Curl Counter
        #Setup status box
        cv2.rectangle(image,(0,0),(225,73),(245,117,16),-1)

        #Rep Data
        cv2.putText(image,'REPS',(15,12),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0),1,cv2.LINE_AA)
        cv2.putText(image ,str(counter) ,(10,60),
                     cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255),2,cv2.LINE_AA)
        
        #Stage Data
        cv2.putText(image,'STAGE',(65,12),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0),1,cv2.LINE_AA)
        cv2.putText(image , stage ,(60,60),
                     cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255),2,cv2.LINE_AA)


        #Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,82), thickness=2, circle_radius=2),
                                 mp_drawing.DrawingSpec(color=(246,67,232), thickness=2, circle_radius=2)
        )  #using utilities of data which we collect after scanning the image continuously

        cv2.imshow('Mediapipe feed',image)  #gives the application window

        if cv2.waitKey(10) & 0xFF == ord('q'):  # waitkey used for exit the application window.
            break
cap.release() # releases the output to the screen
cv2.destroyAllWindows()
```
